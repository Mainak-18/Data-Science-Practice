```python
def check_missing_summary(df):
    missing_data = df.isnull().sum()
    missing_percent = (missing_data / len(df)) * 100
    summary = pd.DataFrame({
        'Missing Values': missing_data,
        'Missing %': missing_percent
    })
    return summary[summary['Missing Values'] > 0].sort_values(by='Missing %', ascending=False)
```

```python
def whisker(col):
    q1,q3 = np.percentile(col,[25,75])
    iqr = q3-q1
    lw = round(q1-1.5*(iqr),2)
    uw = round(q3+1.5*(iqr),2)
    return lw,uw
```

```python
whisker(df['Population'])
```

```python
for i in ['GDP','Population',' thinness  1-19 years',
       ' thinness 5-9 years']:
    lw,uw=whisker(df[i])
    df[i] = np.where(df[i]<lw,lw,df[i])
    df[i] = np.where(df[i]>uw,uw,df[i])
```
