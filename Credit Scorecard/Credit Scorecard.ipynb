```python
def check_missing_df(df):
    null_sum = df.isnull().sum()
    null_percent =( df.isnull().sum()/df.shape[0])*100
    missing_df = pd.DataFrame({
        "Missing Sum":null_sum,
        "Missing %":null_percent
    })
    return missing_df[missing_df['Missing Sum'] > 0].sort_values(by='Missing %', ascending=False)
```
